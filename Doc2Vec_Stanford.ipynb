```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer

from sklearn import metrics


dbow_train_val_test_precise = [[], [], []]
dbow_train_val_test = [[], [], []]

to_class_precise = [0, 1, 2, 3, 4]

security = 1e-4

for sentence in data:
    val = data[sentence]
    tag = val[0]
    split = int(id_to_split[tag]) - 1
    sentiment = float(val[1])
    words = sentence
    
    # Well defined Classes sets
    myC = myClass(words, [tag],
                  to_class_precise[int(sentiment * (len(to_class) - security))])
    dbow_train_val_test_precise[split].append(myC)
    
    # Add in Positive/Negativ sets
    myC = myClass(words, [tag],
                  int(sentiment >= 0.5))
    dbow_train_val_test[split].append(myC)
    
    
# All data
newsgroups_train_data = [s.words for s in dbow_train_val_test[0]]
newsgroups_test_data = [s.words for s in dbow_train_val_test[1]]

newsgroups_train_target = [s.val for s in dbow_train_val_test[0]]
newsgroups_test_target = [s.val for s in dbow_train_val_test[1]]
```

```python

# Create our vectorizer
vectorizer = CountVectorizer()

# Get the training vectors
vectors = vectorizer.fit_transform(newsgroups_train_data)

# Build the classifier
logreg = LogisticRegression(n_jobs=1, C=1e5)

#  Train the classifier
logreg.fit(vectors, newsgroups_train_target)

# Get the test vectors
vectors_test = vectorizer.transform(newsgroups_test_data)

# Predict and score the vectors
pred = logreg.predict(vectors_test)
acc_score = metrics.accuracy_score(newsgroups_test_target, pred)
f1_score = metrics.f1_score(newsgroups_test_target, pred, average='macro')

print('Total accuracy classification score: {}'.format(acc_score))
print('Total F1 classification score: {}'.format(f1_score))
```
